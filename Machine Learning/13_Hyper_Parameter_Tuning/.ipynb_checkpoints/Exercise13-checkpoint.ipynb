```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_digits
import pandas as pd
import numpy as np
```

```python
df = load_digits()
dir(df)
```

```python
pd.DataFrame(df.data)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.data, df.target)
len(x_train)
```

```python
len(x_test)
```

```python
model = {
    'svm':{
        'model' : svm.SVC(gamma='auto'),
        'parameters' :{
            'C' : [1,2,3],
            'kernel' : ['rbf','linear']
        }
    },
    'random_forest_classifier' :{
        'model' : RandomForestClassifier(),
        'parameters':{
            'n_estimators' : [1,2,3]
        }
    },
    'logistic_regression':{
        'model' : LogisticRegression(solver='liblinear',multi_class='auto'),
        'parameters':{
        'C':[1,2,3]
        }
    },
    'GaussianNB' :{
        'model':GaussianNB(),
        'parameters': {}
    },
    'MultinomialNB':{
        'model':MultinomialNB(),
        'parameters':{}
    },
    'DecisionTreeClassifier':{
        'model': DecisionTreeClassifier(),
        'parameters':{
            'criterion' : ['gini','entropy']
        }
    }
}
```

```python
from sklearn.model_selection import GridSearchCV
x=[]
for ml_model, mp in model.items():
    result = GridSearchCV(mp['model'],mp['parameters'], cv = 5,return_train_score = False)
    result.fit(x_train,y_train)
    x.append({
        'model':ml_model,
        'parameters':result.best_params_,
        'score':result.best_score_,
    })
```

```python
verdict = pd.DataFrame(x)
verdict
```
