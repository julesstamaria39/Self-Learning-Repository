```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.columns = ['year','income']
```

```python
%matplotlib inline
plt.scatter(df.year,df.income)
```

```python
df2 = df
df2 = df2.drop('income', axis = 1)
income = df.income
```

```python
reg = linear_model.LinearRegression()
reg.fit(df2,income)
```

```python
reg.predict([[2020]])
```

```python
p = reg.predict(df2)
```

```python
plt.scatter(df.year,df.income)
plt.plot(df2,p,color='red')
```
